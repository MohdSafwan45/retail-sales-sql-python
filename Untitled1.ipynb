```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('retail_sales_analysis/retail_sales_dataset.csv')

print("Dataset loaded successfully!")
print("Shape:", df.shape)
print(df.head(10))

print("=== Dataset Info ===")
print(df.info())

print("\n=== Null Values ===")
print(df.isnull().sum())

print("\n=== Basic Statistics ===")
print(df.describe())

print("\n=== Column Names ===")
print(df.columns.tolist())

print("\n=== Unique Categories ===")
print(df['Product Category'].unique())

df_clean = df.copy()

df_clean['Date'] = pd.to_datetime(df_clean['Date'])

df_clean.drop_duplicates(inplace=True)

df_clean.dropna(inplace=True)

df_clean['Month'] = df_clean['Date'].dt.month
df_clean['Month_Name'] = df_clean['Date'].dt.strftime('%b')
df_clean['Year'] = df_clean['Date'].dt.year

df_clean.columns = df_clean.columns.str.strip().str.replace(' ', '_')

print("Cleaning complete!")
print("Final Shape:", df_clean.shape)
df_clean.head()

conn = sqlite3.connect('retail_sales.db')

df_clean.to_sql('sales', conn, if_exists='replace', index=False)

result = pd.read_sql("SELECT COUNT(*) as total_rows FROM sales", conn)
print("Database ready!")
print(result)

q1 = pd.read_sql("""
SELECT 
    COUNT(*) as total_transactions,
    ROUND(SUM(Total_Amount), 2) as total_revenue,
    ROUND(AVG(Total_Amount), 2) as avg_order_value
FROM sales
""", conn)
print("=== Overall Summary ===")
print(q1)

q2 = pd.read_sql("""
SELECT 
    Product_Category,
    COUNT(*) as transactions,
    ROUND(SUM(Total_Amount), 2) as revenue,
    ROUND(AVG(Total_Amount), 2) as avg_sale
FROM sales
GROUP BY Product_Category
ORDER BY revenue DESC
""", conn)
print("\n=== Revenue by Category ===")
print(q2)

q3 = pd.read_sql("""
SELECT 
    Gender,
    COUNT(*) as customers,
    ROUND(SUM(Total_Amount), 2) as total_spent
FROM sales
GROUP BY Gender
ORDER BY total_spent DESC
""", conn)
print("\n=== Gender-wise Analysis ===")
print(q3)

q4 = pd.read_sql("""
SELECT 
    Month,
    Month_Name,
    COUNT(*) as transactions,
    ROUND(SUM(Total_Amount), 2) as monthly_revenue
FROM sales
GROUP BY Month, Month_Name
ORDER BY Month
""", conn)
print("=== Monthly Revenue Trend ===")
print(q4)

q5 = pd.read_sql("""
SELECT 
    CASE 
        WHEN Age < 25 THEN 'Under 25'
        WHEN Age BETWEEN 25 AND 40 THEN '25-40'
        WHEN Age BETWEEN 41 AND 55 THEN '41-55'
        ELSE '55+'
    END as Age_Group,
    COUNT(*) as customers,
    ROUND(AVG(Total_Amount), 2) as avg_spend
FROM sales
GROUP BY Age_Group
ORDER BY avg_spend DESC
""", conn)
print("\n=== Age Group Analysis ===")
print(q5)

q6 = pd.read_sql("""
SELECT 
    Customer_ID,
    COUNT(*) as purchases,
    ROUND(SUM(Total_Amount), 2) as lifetime_value
FROM sales
GROUP BY Customer_ID
ORDER BY lifetime_value DESC
LIMIT 10
""", conn)
print("\n=== Top 10 Customers ===")
print(q6)

monthly = df_clean.groupby(['Month','Month_Name'])['Total_Amount'].sum().reset_index()
monthly = monthly.sort_values('Month')

plt.figure(figsize=(10, 5))
plt.plot(monthly['Month_Name'], monthly['Total_Amount'], 
         marker='o', color='#1D9E75', linewidth=2.5, markersize=6)
plt.fill_between(monthly['Month_Name'], monthly['Total_Amount'], 
                 alpha=0.1, color='#1D9E75')

plt.title('Monthly Sales Revenue Trend', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Month', fontsize=11)
plt.ylabel('Total Revenue (₹)', fontsize=11)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_trend.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved as monthly_trend.png")

cat_revenue = df_clean.groupby('Product_Category')['Total_Amount'].sum().reset_index()
cat_revenue = cat_revenue.sort_values('Total_Amount', ascending=False)

plt.figure(figsize=(9, 5))
colors = ['#1D9E75','#5DCAA5','#9FE1CB','#E1F5EE']
bars = plt.bar(cat_revenue['Product_Category'], cat_revenue['Total_Amount'], 
               color=colors[:len(cat_revenue)], edgecolor='none')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
             f'₹{bar.get_height():,.0f}', ha='center', va='bottom', 
             fontsize=10, fontweight='bold')

plt.title('Revenue by Product Category', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Category', fontsize=11)
plt.ylabel('Total Revenue (₹)', fontsize=11)
plt.tight_layout()
plt.savefig('category_revenue.png', dpi=150, bbox_inches='tight')
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

gender_data = df_clean.groupby('Gender')['Total_Amount'].sum()
axes[0].pie(gender_data.values, labels=gender_data.index, 
            autopct='%1.1f%%', colors=['#1D9E75','#EEEDFE'],
            startangle=90, textprops={'fontsize':11})
axes[0].set_title('Revenue by Gender', fontsize=13, fontweight='bold')

df_clean['Age_Group'] = pd.cut(df_clean['Age'], 
    bins=[0,25,40,55,100], 
    labels=['Under 25','25-40','41-55','55+'])
age_data = df_clean.groupby('Age_Group')['Total_Amount'].mean().reset_index()
sns.barplot(data=age_data, x='Age_Group', y='Total_Amount', 
            palette='Greens_d', ax=axes[1])
axes[1].set_title('Avg Spend by Age Group', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Average Spend (₹)')

plt.tight_layout()
plt.savefig('demographics_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

```
